### **Seleccionar una fila**


### **Filtrado según un criterio**
Para filtrar el dataframe según una condición, basta con pasar dicha condición como un vector booleano.

Por ejemplo, filtremos el dataframe para mostrar los datos sobre los países asiáticos (AreaName = Asia).

````python
# 1. Crear la serie booleana «condition»
condition = df_can['Continent'] == 'Asia'
print(condition)
````

````python
# 2. Introducir esta condición en el dataFrame
df_can[condition]
````

````python
# Podemos pasar varios criterios en la misma línea.
# Filtremos por AreaName = Asia y RegName = Asia Meridional

df_can[(df_can['Continent']=='Asia') & (df_can['Region']=='Southern Asia')]

# Nota: al utilizar los operadores «and» y «or», pandas requiere que utilicemos «&» y «|» en lugar de «and» y «or».
# No olvides encerrar las dos condiciones entre paréntesis.
````

### **Ordenación de valores de un DataFrame o una serie** <br>
La función `sort_values()` se utiliza para ordenar un DataFrame o una serie en función de una o varias columnas. <br>Especificar la columna o columnas por las que se desea ordenar y el orden (ascendente o descendente). A continuación se muestra la sintaxis para utilizarla:<br>

```df.sort_values(col_name, axis=0, ascending=True, inplace=False, ignore_index=False)```<br>

col_name: la(s) columna(s) por la(s) que se va a ordenar. <br>
axis: eje a lo largo del cual se va a ordenar. 0 para ordenar por filas (por defecto) y 1 para ordenar por columnas.<br>
ascending: para ordenar en orden ascendente (True, por defecto) o descendente (False).<br>
inplace: para realizar la operación de ordenación in situ (True) o devolver una copia ordenada (False, por defecto).<br>
ignore_index: para restablecer el índice tras la ordenación (True) o mantener los valores originales del índice (False, por defecto).<br>

Ordenemos el dataframe df_can por la columna «Total», en orden descendente, para averiguar cuáles son los cinco países que más han contribuido a la inmigración en Canadá.

````python
df_can.sort_values(by='Total', axis=0, ascending=False, inplace=False, ignore_index=False).head(5)
````
